```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Data = pd.read_csv("DataSet/Churn_Modelling.csv")
Data.head()
```

```python
Data.shape
```

```python
X = Data.iloc[:,3:13].values
(pd.DataFrame(X)).head(10)
```

```python
Y = Data.iloc[:,13].values
(pd.DataFrame(Y)).head(10)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
labelencoder_x1 = LabelEncoder()
X[:,1] = labelencoder_x1.fit_transform(X[:,1])

LabelEncoder_x2 = LabelEncoder()
X[:,2] = LabelEncoder_x2.fit_transform(X[:,2])
```

```python
pd.DataFrame(X[:,1]).head()
```

```python
pd.DataFrame(X[:,2]).head()
```

```python
(pd.DataFrame(X)).head()
```

```python
OneHotEncoder = OneHotEncoder(categorical_features=[1])
X = OneHotEncoder.fit_transform(X).toarray()
```

```python
X
```

```python
X=X[:,1:]
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_tr,X_ts,Y_tr,Y_ts = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_x = StandardScaler()
X_tr = sc_x.fit_transform(X_tr)
X_ts = sc_x.transform(X_ts)
```

```python
(pd.DataFrame(X_tr)).head()
```

```python
(pd.DataFrame(X_ts)).head()
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
Classifier = Sequential()
Classifier.add(Dense(units=6, kernel_initializer="uniform", activation="relu",input_dim=11))
```

```python
#units=n es el num de neuronas
Classifier.add(Dense(units=6, kernel_initializer="uniform", activation="relu"))
```

```python
Classifier.add(Dense(units=1, kernel_initializer="uniform", activation="sigmoid"))
```

```python
Classifier.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
Classifier.fit(X_tr,Y_tr,batch_size=10,epochs=100)
#batch size es el tamaño de bloque el cual agarra
```

```python
Y_pr = Classifier.predict(X_ts)
Y_pr
```

```python
Y_pr = (Y_pr > 0.5)
pd.DataFrame(Y_pr) 
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(Y_pr,Y_ts)
cm
```
